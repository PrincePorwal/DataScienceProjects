```python
Q1 = data['campaign'].quantile(0.25)
Q3 = data['campaign'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

upper_bound, lower_bound
```

```python
data = data[(data['campaign'] <= upper_bound) & (data['campaign'] >= lower_bound)]
```

```python
Q1 = data['duration'].quantile(0.25)
Q3 = data['duration'].quantile(0.75)
IQR = Q3 - Q1

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

upper_bound, lower_bound
```

```python
data = data[(data['duration'] <= upper_bound) & (data['duration'] >= lower_bound)]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data.dtypes
```

```python
# Label encoding for categorical variables
label_encoder = LabelEncoder()
data['job'] = label_encoder.fit_transform(data['job'])
data['marital'] = label_encoder.fit_transform(data['marital'])
data['education'] = label_encoder.fit_transform(data['education'])
data['default'] = label_encoder.fit_transform(data['default'])
data['housing'] = label_encoder.fit_transform(data['housing'])
data['loan'] = label_encoder.fit_transform(data['loan'])
data['contact'] = label_encoder.fit_transform(data['contact'])
data['month'] = label_encoder.fit_transform(data['month'])
data['day_of_week'] = label_encoder.fit_transform(data['day_of_week'])
data['poutcome'] = label_encoder.fit_transform(data['poutcome'])
data['y'] = label_encoder.fit_transform(data['y'])

```

```python
plt.figure(figsize=(15,15))
cor = data.corr()
sns.heatmap(cor, annot=True, cmap= "Greens")
plt.show()
```

'y' is highly co-related to duration, previous and poutcome

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# feature Selection
X = data.drop(columns=['y', 'emp.var.rate', 'cons.conf.idx']) # dropped emp.var.rate and cons.conf.idx because they had negative values(selectkbest was no working because of that)
y = data['y']
X_scaled = mms.fit_transform(X) # Scaling the data
X_new = SelectKBest(chi2, k=10).fit_transform(X_scaled, y) # selecting top 10 features
```

```python
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score , roc_auc_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2, random_state=42)
```

```python
logistic = LogisticRegression(solver = 'saga', penalty= 'l1', C = 100)
NB = GaussianNB()
xgb = XGBClassifier(n_estimators= 100, max_depth= 3, learning_rate= 0.3)
randomforest = RandomForestClassifier(n_estimators= 100, min_samples_split= 5, min_samples_leaf= 1, max_depth= 10, bootstrap= True)
decisiontree= DecisionTreeClassifier(min_samples_split= 2, min_samples_leaf= 4, max_depth= 10)
```

```python
# Using randomsearchcv to get best params for hypertuning
param_grid_logistic = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga']
}
random_search = RandomizedSearchCV(logistic, param_distributions=param_grid_logistic, n_iter=10, cv=5, verbose=1, n_jobs=-1)

random_search.fit(X_train, y_train)

best_params = random_search.best_params_
best_score = random_search.best_score_

print(best_params)
print(best_score)
```

```python
logistic.fit(X_train, y_train)
```

```python
y_pred_logistic = logistic.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_logistic))
print(roc_auc_score(y_test, y_pred_logistic))
```

roc score(without outlier removal): 0.6836

roc score(after outlier removal): 0.698

After HyperParameter Tuning:

roc score: 0.704

```python
NB.fit(X_train, y_train)
```

```python
y_pred_NB = NB.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_NB))
print(roc_auc_score(y_test, y_pred_NB))
```

roc score(without outlier removal): 0.7379

roc score(after outlier removal): 0.752

Gaussian Naive Bayes dont have any hyperparameters to tune

```python
# Define the parameter grid for XGBoost
param_grid_xgb = {
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [3, 4, 5, 6],

}

# Create a RandomizedSearchCV instance
random_search = RandomizedSearchCV(xgb, param_distributions=param_grid_xgb, n_iter=10, cv=5, verbose=1, n_jobs=-1)

# Fit the model to your data
random_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = random_search.best_params_
best_score = random_search.best_score_

print(best_params)
print(best_score)
```

```python
xgb.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb.predict(X_test)
print(classification_report(y_test, y_pred_xgb))
print(roc_auc_score(y_test, y_pred_xgb))
```
